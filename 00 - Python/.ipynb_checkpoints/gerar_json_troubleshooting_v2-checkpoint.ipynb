```python
r"""
Gerador de JSON - Wolf Troubleshooting
Versao: v4

Objetivo:
- Ler a planilha pai do Troubleshooting.
- Gerar o arquivo dados_troubleshooting.json para ser consumido pelo ambiente web offline.
- Gerar o schema_troubleshooting.json como documentacao tecnica basica da estrutura.

Mudanca principal da v4:
- O campo oficial de correlacao entre abas passa a ser sempre o CODIGO da falha/alerta.
- A aba 01_Troubleshooting nao precisa mais ter coluna ID.
- As abas 02_Diagnostico_Guiado, 03_Anexos e 04_Revisoes devem referenciar os itens pelos codigos,
  por exemplo: FST003, FST001, ASP007 etc.

Estrutura de pastas esperada no Windows:
C:\Users\wesley\Documents\13 - TroubleShooting\
├── 00 - Python\
│   └── gerar_json_troubleshooting.py
├── 01 - DB\
│   └── Base_Pai_Troubleshooting_Alertas_Falhas_MVP.xlsx
└── 99 - Saidas JSON\
    ├── dados_troubleshooting.json
    ├── schema_troubleshooting.json
    └── relatorio_geracao_json.txt

Dependencia:
pip install openpyxl
"""

from __future__ import annotations

from collections import defaultdict
from datetime import datetime, date
from pathlib import Path
import argparse
import hashlib
import json
import re
import sys
from typing import Any

try:
    from openpyxl import load_workbook
except ImportError as exc:
    raise SystemExit(
        "ERRO: a biblioteca openpyxl nao esta instalada.\n"
        "Instale com: pip install openpyxl"
    ) from exc


# =============================================================================
# CONFIGURACAO PADRAO DE PASTAS
# =============================================================================

PROJECT_ROOT = Path.home() / "Documents" / "13 - TroubleShooting"
DEFAULT_EXCEL_PATH = PROJECT_ROOT / "01 - DB" / "Base_Pai_Troubleshooting_Alertas_Falhas_MVP.xlsx"
DEFAULT_OUTPUT_DIR = PROJECT_ROOT / "99 - Saidas JSON"
DEFAULT_JSON_NAME = "dados_troubleshooting.json"
DEFAULT_SCHEMA_NAME = "schema_troubleshooting.json"
DEFAULT_REPORT_NAME = "relatorio_geracao_json.txt"


# =============================================================================
# MAPEAMENTO DOS CABECALHOS DA PLANILHA PARA CHAVES JSON
# =============================================================================

HEADER_MAP = {
    # Aba 01_Troubleshooting
    "ID": "id_antigo",
    "Código": "codigo",
    "Codigo": "codigo",
    "Código Falha/Alerta": "codigo",
    "Codigo Falha/Alerta": "codigo",
    "Código do Item": "codigo",
    "Codigo do Item": "codigo",
    "Tipo": "tipo",
    "Sistema": "sistema",
    "Subsistema": "subsistema",
    "Severidade": "severidade",
    "Prioridade": "prioridade",
    "Status": "status",
    "Mensagem na IHM / Alerta": "mensagem_ihm_alerta",
    "Descrição resumida": "descricao_resumida",
    "Descricao resumida": "descricao_resumida",
    "Causa": "causa",
    "Efeito na máquina/operação": "efeito_maquina_operacao",
    "Efeito na maquina/operacao": "efeito_maquina_operacao",
    "Solução": "solucao",
    "Solucao": "solucao",
    "Validação após solução": "validacao_apos_solucao",
    "Validacao apos solucao": "validacao_apos_solucao",
    "Origem": "origem",
    "Variável / Código relacionado": "variavel_codigo_relacionado",
    "Variavel / Codigo relacionado": "variavel_codigo_relacionado",
    "Ferramentas necessárias": "ferramentas_necessarias",
    "Ferramentas necessarias": "ferramentas_necessarias",
    "Observações": "observacoes",
    "Observacoes": "observacoes",
    "Responsável": "responsavel",
    "Responsavel": "responsavel",
    "Data revisão": "data_revisao",
    "Data revisao": "data_revisao",
    "Versão": "versao",
    "Versao": "versao",

    # Aba 02_Diagnostico_Guiado
    "ID Passo": "id_passo",
    "Código Falha/Alerta": "codigo_item",
    "Codigo Falha/Alerta": "codigo_item",
    "Código do Item": "codigo_item",
    "Codigo do Item": "codigo_item",
    "Código Referência": "codigo_item",
    "Codigo Referencia": "codigo_item",
    "ID Falha/Alerta": "codigo_item",  # Mantido por compatibilidade, mas agora deve conter o codigo.
    "Ordem": "ordem",
    "Pergunta/Teste": "pergunta_teste",
    "Resposta esperada": "resposta_esperada",
    "Se OK": "se_ok",
    "Se não OK": "se_nao_ok",
    "Se nao OK": "se_nao_ok",
    "Ferramentas": "ferramentas",

    # Aba 03_Anexos
    "ID Anexo": "id_anexo",
    "Título": "titulo",
    "Titulo": "titulo",
    "Arquivo/Caminho relativo": "arquivo_caminho_relativo",
    "Descrição": "descricao",
    "Descricao": "descricao",

    # Aba 04_Revisoes
    "Data": "data",
    "Tipo alteração": "tipo_alteracao",
    "Tipo alteracao": "tipo_alteracao",
    "Itens alterados": "itens_alterados",
    "Status publicação": "status_publicacao",
    "Status publicacao": "status_publicacao",
}

REQUIRED_SHEETS = [
    "01_Troubleshooting",
    "02_Diagnostico_Guiado",
    "03_Anexos",
    "04_Revisoes",
    "Listas",
]


# =============================================================================
# FUNCOES UTILITARIAS
# =============================================================================

def slug(value: object) -> str:
    """Converte um texto em uma chave simples para JSON."""
    text = str(value).strip().lower()
    replacements = {
        "á": "a", "à": "a", "â": "a", "ã": "a", "ä": "a",
        "é": "e", "ê": "e", "è": "e", "ë": "e",
        "í": "i", "ì": "i", "î": "i", "ï": "i",
        "ó": "o", "ô": "o", "õ": "o", "ò": "o", "ö": "o",
        "ú": "u", "ù": "u", "û": "u", "ü": "u",
        "ç": "c",
    }
    for source, target in replacements.items():
        text = text.replace(source, target)
    text = re.sub(r"[^a-z0-9]+", "_", text).strip("_")
    return text or "campo"


def cell_value(value: object) -> object:
    """Normaliza valores vindos do Excel para tipos compativeis com JSON."""
    if isinstance(value, datetime):
        return value.date().isoformat()
    if isinstance(value, date):
        return value.isoformat()
    if isinstance(value, str):
        cleaned = value.strip()
        return cleaned if cleaned else None
    return value


def normalize_code(value: object) -> str:
    """Normaliza codigos para chave de relacionamento entre abas."""
    if value is None:
        return ""
    text = str(value).strip()
    if not text:
        return ""
    # Evita transformar codigos textuais, mas normaliza espacamentos e caixa.
    text = re.sub(r"\s+", "", text)
    return text.upper()


def to_int_if_possible(value: object, default: int = 9999) -> int:
    try:
        return int(value)  # type: ignore[arg-type]
    except (TypeError, ValueError):
        return default


def first_nonempty(values: list[object], default: object = None) -> object:
    for value in values:
        if value not in (None, ""):
            return value
    return default


def get_reference_code(row: dict[str, Any]) -> str:
    """
    Retorna o codigo usado para relacionar uma linha secundaria ao item principal.

    A partir da v4, o relacionamento oficial e sempre pelo codigo da falha/alerta.
    Esta funcao aceita varios nomes de coluna para manter compatibilidade com versoes antigas da planilha.
    """
    possible_keys = [
        "codigo_item",
        "codigo",
        "id_item",      # compatibilidade com planilhas antigas, caso ainda exista
        "id_antigo",    # nao recomendado
        "item_codigo",
        "codigo_falha_alerta",
        "codigo_referencia",
    ]
    for key in possible_keys:
        code = normalize_code(row.get(key))
        if code:
            return code
    return ""


def clean_secondary_row(row: dict[str, Any]) -> dict[str, Any]:
    """Remove chaves internas redundantes das linhas secundarias."""
    cleaned = dict(row)
    # Mantem codigo_item como campo explicito para rastreabilidade no JSON.
    if "codigo" in cleaned and "codigo_item" not in cleaned:
        cleaned["codigo_item"] = cleaned["codigo"]
    for key in ["id_antigo", "id_item"]:
        cleaned.pop(key, None)
    if "codigo_item" in cleaned:
        cleaned["codigo_item"] = normalize_code(cleaned["codigo_item"])
    return cleaned


def read_sheet_rows(workbook, sheet_name: str) -> list[dict[str, Any]]:
    """Le uma aba usando a primeira linha como cabecalho."""
    ws = workbook[sheet_name]
    headers = [cell_value(cell.value) for cell in ws[1]]

    keys: list[str] = []
    seen: dict[str, int] = {}

    for header in headers:
        if header is None:
            key = "campo_vazio"
        else:
            header_text = str(header).strip()
            key = HEADER_MAP.get(header_text, slug(header_text))

        if key in seen:
            seen[key] += 1
            key = f"{key}_{seen[key]}"
        else:
            seen[key] = 1
        keys.append(key)

    rows: list[dict[str, Any]] = []
    for row in ws.iter_rows(min_row=2, values_only=True):
        values = [cell_value(value) for value in row[: len(keys)]]
        if not any(value is not None for value in values):
            continue
        rows.append({key: value for key, value in zip(keys, values) if value is not None})
    return rows


def read_lists(workbook) -> dict[str, list[object]]:
    """Le a aba Listas preservando cada coluna como uma lista auxiliar."""
    ws = workbook["Listas"]
    headers = [cell_value(cell.value) for cell in ws[1]]
    lists: dict[str, list[object]] = {}

    for col_idx, header in enumerate(headers, start=1):
        if not header:
            continue
        key = slug(header)
        values: list[object] = []
        for row_idx in range(2, ws.max_row + 1):
            value = cell_value(ws.cell(row_idx, col_idx).value)
            if value is not None and value not in values:
                values.append(value)
        lists[key] = values
    return lists


def validate_workbook(workbook) -> list[str]:
    """Valida estrutura minima da planilha."""
    warnings: list[str] = []
    for sheet in REQUIRED_SHEETS:
        if sheet not in workbook.sheetnames:
            warnings.append(f"Aba obrigatoria ausente: {sheet}")
    return warnings


def build_payload(excel_path: Path) -> tuple[dict[str, Any], list[str]]:
    """Monta o corpo completo do JSON a partir da planilha pai."""
    workbook = load_workbook(excel_path, data_only=True)
    warnings = validate_workbook(workbook)
    if warnings:
        raise ValueError("A planilha nao possui a estrutura minima esperada:\n" + "\n".join(warnings))

    troubleshooting = read_sheet_rows(workbook, "01_Troubleshooting")
    diagnostics = read_sheet_rows(workbook, "02_Diagnostico_Guiado")
    attachments = read_sheet_rows(workbook, "03_Anexos")
    revisions = read_sheet_rows(workbook, "04_Revisoes")
    lists = read_lists(workbook)

    # Normaliza a chave oficial da tabela principal: codigo.
    items_by_code: dict[str, dict[str, Any]] = {}
    duplicated_codes: set[str] = set()
    for item in troubleshooting:
        code = normalize_code(item.get("codigo"))
        if not code:
            warnings.append("Item na aba 01_Troubleshooting sem codigo. Linha ignorada na correlacao.")
            continue
        item["codigo"] = code
        # A partir da v4, id_item e o proprio codigo para facilitar o consumo pelo HTML.
        item["id_item"] = code
        if code in items_by_code:
            duplicated_codes.add(code)
        items_by_code[code] = item

    if duplicated_codes:
        warnings.append("Codigos duplicados encontrados na aba 01_Troubleshooting: " + ", ".join(sorted(duplicated_codes)))

    # Agrupa diagnosticos por codigo da falha/alerta.
    diagnostics_by_code: dict[str, list[dict[str, Any]]] = defaultdict(list)
    for diagnostic in diagnostics:
        code = get_reference_code(diagnostic)
        if code:
            cleaned = clean_secondary_row(diagnostic)
            cleaned["codigo_item"] = code
            diagnostics_by_code[code].append(cleaned)
    for value in diagnostics_by_code.values():
        value.sort(key=lambda row: to_int_if_possible(row.get("ordem")))

    # Agrupa anexos por codigo da falha/alerta.
    attachments_by_code: dict[str, list[dict[str, Any]]] = defaultdict(list)
    for attachment in attachments:
        code = get_reference_code(attachment)
        if code:
            cleaned = clean_secondary_row(attachment)
            cleaned["codigo_item"] = code
            attachments_by_code[code].append(cleaned)
    for value in attachments_by_code.values():
        value.sort(key=lambda row: to_int_if_possible(row.get("ordem")))

    # Agrupa revisoes. Uma revisao pode referenciar varios codigos separados por ; ou ,.
    revisions_by_code: dict[str, list[dict[str, Any]]] = defaultdict(list)
    for revision in revisions:
        raw_ids = revision.get("itens_alterados") or revision.get("codigo_item") or revision.get("codigo")
        if not raw_ids:
            continue
        codes = [normalize_code(part) for part in re.split(r"[;,]", str(raw_ids)) if normalize_code(part)]
        for code in codes:
            revisions_by_code[code].append(dict(revision))

    # Validacao de orfaos, agora por codigo.
    main_codes = set(items_by_code.keys())
    for diagnostic in diagnostics:
        code = get_reference_code(diagnostic)
        if code and code not in main_codes:
            warnings.append(f"Diagnostico orfao: {diagnostic.get('id_passo')} referencia codigo inexistente {code}")
        elif not code and any(value is not None for value in diagnostic.values()):
            warnings.append(f"Diagnostico sem codigo de referencia: {diagnostic.get('id_passo')}")

    for attachment in attachments:
        code = get_reference_code(attachment)
        if code and code not in main_codes:
            warnings.append(f"Anexo orfao: {attachment.get('id_anexo')} referencia codigo inexistente {code}")
        elif not code and any(value is not None for value in attachment.values()):
            warnings.append(f"Anexo sem codigo de referencia: {attachment.get('id_anexo')}")

    for revision in revisions:
        raw_ids = revision.get("itens_alterados") or revision.get("codigo_item") or revision.get("codigo")
        if not raw_ids:
            continue
        codes = [normalize_code(part) for part in re.split(r"[;,]", str(raw_ids)) if normalize_code(part)]
        for code in codes:
            if code not in main_codes:
                warnings.append(f"Revisao referencia codigo inexistente {code}")

    items: list[dict[str, Any]] = []

    for item in troubleshooting:
        code = normalize_code(item.get("codigo"))
        if not code:
            continue

        search_parts: list[str] = []
        search_fields = [
            "id_item",
            "codigo",
            "tipo",
            "sistema",
            "subsistema",
            "severidade",
            "prioridade",
            "status",
            "mensagem_ihm_alerta",
            "descricao_resumida",
            "causa",
            "efeito_maquina_operacao",
            "solucao",
            "validacao_apos_solucao",
            "origem",
            "variavel_codigo_relacionado",
            "ferramentas_necessarias",
            "observacoes",
        ]
        for field in search_fields:
            if item.get(field):
                search_parts.append(str(item[field]))

        for diagnostic in diagnostics_by_code.get(code, []):
            for field in ["pergunta_teste", "resposta_esperada", "se_ok", "se_nao_ok", "ferramentas"]:
                if diagnostic.get(field):
                    search_parts.append(str(diagnostic[field]))

        for attachment in attachments_by_code.get(code, []):
            for field in ["titulo", "descricao", "arquivo_caminho_relativo", "tipo_anexo"]:
                if attachment.get(field):
                    search_parts.append(str(attachment[field]))

        entry = {
            "id_item": code,
            "codigo": code,
            "identificacao": {
                "codigo": code,
                "tipo": item.get("tipo"),
                "sistema": item.get("sistema"),
                "subsistema": item.get("subsistema"),
                "origem": item.get("origem"),
                "variavel_codigo_relacionado": item.get("variavel_codigo_relacionado"),
            },
            "classificacao": {
                "severidade": item.get("severidade"),
                "prioridade": item.get("prioridade"),
                "status": item.get("status"),
            },
            "conteudo": {
                "mensagem_ihm_alerta": item.get("mensagem_ihm_alerta"),
                "descricao_resumida": item.get("descricao_resumida"),
                "causa": item.get("causa"),
                "efeito_maquina_operacao": item.get("efeito_maquina_operacao"),
                "solucao": item.get("solucao"),
                "validacao_apos_solucao": item.get("validacao_apos_solucao"),
                "ferramentas_necessarias": item.get("ferramentas_necessarias"),
                "observacoes": item.get("observacoes"),
            },
            "controle": {
                "responsavel": item.get("responsavel"),
                "data_revisao": item.get("data_revisao"),
                "versao": item.get("versao"),
            },
            "diagnostico_guiado": diagnostics_by_code.get(code, []),
            "anexos": attachments_by_code.get(code, []),
            "revisoes": revisions_by_code.get(code, []),
            "search_text": " | ".join(search_parts),
        }
        items.append(entry)

    versions = [item.get("versao") for item in troubleshooting if item.get("versao")]
    latest_version = first_nonempty(list(reversed(versions)), "0.0.0")
    revision_dates = [item.get("data_revisao") for item in troubleshooting if item.get("data_revisao")]
    latest_date = max(revision_dates) if revision_dates else None

    payload: dict[str, Any] = {
        "metadata": {
            "produto": "Wolf Troubleshooting de Alertas e Falhas",
            "formato": "dados_troubleshooting",
            "schema_version": "1.0.0",
            "versao_base": latest_version,
            "data_referencia_base": latest_date,
            "data_geracao_json": datetime.now().isoformat(timespec="seconds"),
            "arquivo_fonte": str(excel_path),
            "chave_correlacao": "codigo",
            "observacao": "Arquivo JSON gerado a partir da planilha pai. A planilha .xlsx permanece como documento primordial da base. A correlacao entre abas usa o codigo da falha/alerta.",
            "totais": {
                "itens_troubleshooting": len(items),
                "passos_diagnostico": len(diagnostics),
                "anexos": len(attachments),
                "revisoes": len(revisions),
            },
            "avisos_validacao": warnings,
        },
        "configuracao_interface": {
            "campos_pesquisa": [
                "codigo",
                "mensagem_ihm_alerta",
                "descricao_resumida",
                "causa",
                "efeito_maquina_operacao",
                "solucao",
                "sistema",
                "subsistema",
                "origem",
                "variavel_codigo_relacionado",
                "search_text",
            ],
            "filtros_principais": ["tipo", "sistema", "severidade", "prioridade", "status", "origem"],
            "ordenacao_padrao": ["sistema", "severidade", "codigo"],
            "status_publicaveis_sugeridos": ["Homologado", "Publicado"],
            "modo_operacao": "offline",
        },
        "listas": lists,
        "itens": items,
    }

    text_without_hash = json.dumps(payload, ensure_ascii=False, indent=2)
    payload["metadata"]["sha256_conteudo"] = hashlib.sha256(text_without_hash.encode("utf-8")).hexdigest()
    return payload, warnings


def build_schema() -> dict[str, Any]:
    """Cria um schema JSON simplificado para documentacao e validacao futura."""
    return {
        "$schema": "https://json-schema.org/draft/2020-12/schema",
        "title": "Schema - Wolf Troubleshooting de Alertas e Falhas",
        "type": "object",
        "required": ["metadata", "configuracao_interface", "listas", "itens"],
        "properties": {
            "metadata": {"type": "object"},
            "configuracao_interface": {"type": "object"},
            "listas": {"type": "object"},
            "itens": {
                "type": "array",
                "items": {
                    "type": "object",
                    "required": [
                        "id_item",
                        "codigo",
                        "identificacao",
                        "classificacao",
                        "conteudo",
                        "diagnostico_guiado",
                        "anexos",
                    ],
                    "properties": {
                        "id_item": {"type": ["string", "null"]},
                        "codigo": {"type": ["string", "null"]},
                        "identificacao": {"type": "object"},
                        "classificacao": {"type": "object"},
                        "conteudo": {"type": "object"},
                        "controle": {"type": "object"},
                        "diagnostico_guiado": {"type": "array"},
                        "anexos": {"type": "array"},
                        "revisoes": {"type": "array"},
                        "search_text": {"type": "string"},
                    },
                },
            },
        },
    }


def save_outputs(payload: dict[str, Any], output_dir: Path) -> tuple[Path, Path, Path]:
    """Salva JSON, schema e relatorio na pasta de saida."""
    output_dir.mkdir(parents=True, exist_ok=True)

    json_path = output_dir / DEFAULT_JSON_NAME
    schema_path = output_dir / DEFAULT_SCHEMA_NAME
    report_path = output_dir / DEFAULT_REPORT_NAME

    json_path.write_text(json.dumps(payload, ensure_ascii=False, indent=2), encoding="utf-8")
    schema_path.write_text(json.dumps(build_schema(), ensure_ascii=False, indent=2), encoding="utf-8")

    metadata = payload["metadata"]
    report_lines = [
        "RELATORIO DE GERACAO DO JSON - WOLF TROUBLESHOOTING",
        "=" * 62,
        f"Data/hora geracao: {metadata.get('data_geracao_json')}",
        f"Arquivo fonte: {metadata.get('arquivo_fonte')}",
        f"Versao base: {metadata.get('versao_base')}",
        f"Data referencia base: {metadata.get('data_referencia_base')}",
        f"Chave de correlacao entre abas: {metadata.get('chave_correlacao')}",
        "",
        "Totais:",
        f"- Itens troubleshooting: {metadata['totais'].get('itens_troubleshooting')}",
        f"- Passos diagnostico: {metadata['totais'].get('passos_diagnostico')}",
        f"- Anexos: {metadata['totais'].get('anexos')}",
        f"- Revisoes: {metadata['totais'].get('revisoes')}",
        "",
        f"SHA-256 conteudo: {metadata.get('sha256_conteudo')}",
        "",
        "Avisos de validacao:",
    ]
    warnings = metadata.get("avisos_validacao") or []
    if warnings:
        report_lines.extend([f"- {warning}" for warning in warnings])
    else:
        report_lines.append("- Nenhum aviso encontrado.")

    report_path.write_text("\n".join(report_lines), encoding="utf-8")
    return json_path, schema_path, report_path


def parse_args() -> argparse.Namespace:
    parser = argparse.ArgumentParser(description="Gera JSON a partir da base pai do Wolf Troubleshooting.")
    parser.add_argument(
        "--excel",
        type=Path,
        default=DEFAULT_EXCEL_PATH,
        help=f"Caminho da planilha pai. Padrao: {DEFAULT_EXCEL_PATH}",
    )
    parser.add_argument(
        "--saida",
        type=Path,
        default=DEFAULT_OUTPUT_DIR,
        help=f"Pasta de saida dos JSONs. Padrao: {DEFAULT_OUTPUT_DIR}",
    )

    # Ambientes como Jupyter/Spyder/IPython podem enviar argumentos extras automaticamente.
    # parse_known_args() ignora esses argumentos e evita erro com parametro -f kernel.json.
    args, unknown_args = parser.parse_known_args()

    if unknown_args:
        print("Aviso: argumentos externos ignorados:", " ".join(str(arg) for arg in unknown_args))

    return args


def main() -> int:
    args = parse_args()
    excel_path: Path = args.excel
    output_dir: Path = args.saida

    print("Wolf Troubleshooting - Gerador de JSON")
    print("-" * 45)
    print(f"Planilha fonte: {excel_path}")
    print(f"Pasta saida:    {output_dir}")
    print("Chave correlacao: codigo")

    if not excel_path.exists():
        print("\nERRO: planilha pai nao encontrada.")
        print("Verifique se o arquivo esta exatamente neste caminho:")
        print(excel_path)
        return 1

    try:
        payload, warnings = build_payload(excel_path)
        json_path, schema_path, report_path = save_outputs(payload, output_dir)
    except Exception as exc:
        print(f"\nERRO durante a geracao: {exc}")
        return 1

    print("\nArquivos gerados com sucesso:")
    print(f"- {json_path}")
    print(f"- {schema_path}")
    print(f"- {report_path}")
    print("\nResumo:")
    print(f"- Itens: {payload['metadata']['totais']['itens_troubleshooting']}")
    print(f"- Passos diagnostico: {payload['metadata']['totais']['passos_diagnostico']}")
    print(f"- Anexos: {payload['metadata']['totais']['anexos']}")
    print(f"- Revisoes: {payload['metadata']['totais']['revisoes']}")

    if warnings:
        print("\nAvisos de validacao encontrados:")
        for warning in warnings:
            print(f"- {warning}")
    else:
        print("\nNenhum aviso de validacao encontrado.")

    return 0


if __name__ == "__main__":
    sys.exit(main())
```
